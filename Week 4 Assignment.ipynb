```python
import numpy as np 
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', False)
```

```python
column_names = ['Postal_Code','Borough','Neighborhood']
pd_Neighbor=pd.DataFrame(columns=column_names)
```

```python
pd_Neighbor
```

```python
from urllib.request import urlopen
wiki = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urlopen(wiki)
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "lxml")
print(soup.prettify())
```

```python
Table1=soup.find('table', class_='wikitable sortable')
Table1
```

```python
A=[]
B=[]
C=[]

for row in Table1.findAll("tr"):
    cells = row.findAll('td')
    if len(cells)==3: #Only extract table body not heading
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))

        
#Adding Data to our DataFrame
pd_Neighbor['Postal_Code']=A
pd_Neighbor['Borough']=B
pd_Neighbor['Neighborhood']=C

pd_Neighbor
```
